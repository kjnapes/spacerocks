```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
```

```python
def stumpff_c(z):
    if z > 0:
        return (1 - np.cos(np.sqrt(z))) / z
    elif z < 0:
        return (np.cosh(np.sqrt(-z)) - 1) / -z
    else:
        return 1/2
    
def stumpff_s(z):
    if z > 0:
        return (np.sqrt(z) - np.sin(np.sqrt(z))) / np.sqrt(z**3)
    elif z < 0:
        return (np.sinh(np.sqrt(-z)) - np.sqrt(-z)) / np.sqrt((-z)**3)
    else:
        return 1/6
```

```python
xs = np.linspace(0, 500, 1000)
ys = [stumpff_s(x) for x in xs]
yc = [stumpff_c(x) for x in xs]

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(xs, ys, label=r'$S(z)$', color='crimson')
ax.plot(xs, yc, label=r'$C(z)$', color='royalblue')  

ax.set_xlim(0, 500)
ax.set_ylim(bottom=0)
ax.set_yscale('symlog', linthresh=0.001)
ax.legend(fontsize=14)
ax.set_xlabel('z', fontsize=14, labelpad=10)
ax.set_ylabel('value', fontsize=14, labelpad=30)
```
